```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
x=pd.read_csv('/content/diabetes.tsv',sep='\t')
```

```python
x
```

```python
#summary of our data :
x.info()
```

```python
print(x)
```

```python
#statistical values for columns:
x.describe()
```

```python
#Calculating mean,min and max of BMI:
mean_bmi=x['BMI'].mean()
min_bmi=x['BMI'].min()
max_bmi=x['BMI'].max()
print(mean_bmi)
print(min_bmi)
print(max_bmi)
```

```python
#Calculating mean ,std, min and max of BP:
mean_bp=x['BP'].mean()
std_bp=x['BP'].std()
min_bp=x['BP'].min()
max_bp=x['BP'].max()
print("Mean BP is :"+str(mean_bp))
print("Std BP is :"+str(std_bp))
print("Min BP is :"+str(min_bp))
print("Max BP is :"+str(max_bp))
```

```python
plt.boxplot(x['BMI'],vert=False)
plt.show()
```

```python
plt.figure(figsize = (8,2))
plt.boxplot(x['BMI'],vert=False)
plt.show()
```

```python
plt.figure(figsize = (8,2))
plt.boxplot(x['BMI'],vert=False,showmeans=True)
plt.grid()
plt.title("This  is a BoxPlot for the BMI of patients:")
plt.xlabel("BMI of Patients")

plt.show()
```

```python
plt.figure(figsize = (8,2))
plt.boxplot(x['AGE'],vert=False,showmeans=True)
plt.grid()
plt.title("This  is a BoxPlot for the AGE of patients:")
plt.xlabel("AGE of Patients")

plt.show()
```

```python
plt.figure(figsize = (8,2))
plt.boxplot(x['AGE'],vert=False,showmeans=True)
plt.grid()
plt.grid(axis='y')
plt.title("This  is a BoxPlot for the AGE of patients:")
plt.xlabel("AGE of Patients")
plt.ylabel("AGE")
plt.show()
```
