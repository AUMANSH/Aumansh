```python
import seaborn as sns #advanced visualisation
```

```python
sns.histplot(x['age'])
plt.title("Age Distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
sns.histplot(x['age'])
plt.title("Age Distribution" ,size =12)
plt.xlabel("age", size =12)
plt.ylabel("frequency", size =12)
plt.show()
```

```python
sns.histplot(x['age'])
plt.title("Age Distribution" ,size =12)
plt.xlabel("age", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
sns.histplot(x['bmi'])
plt.title("BMI Distribution" ,size =12)
plt.xlabel("bmi", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
sns.histplot(x['bmi'] , color="g")
plt.title("BMI Distribution" ,size =12)
plt.xlabel("bmi", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
#countplot(in sns) for sex
sns.countplot(x['sex'])
plt.title("Sex Distribution" ,size =12)
plt.xlabel("sex", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
#countplot(in sns) for smoker
sns.countplot(x['smoker'])
plt.title("Smoker Distribution" ,size =12)
plt.xlabel("smoker", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
#children distribution :0,1,2,3,4,5
sns.countplot(x['children'])
plt.title("Children Distribution" ,size =12)
plt.xlabel("children", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
#region distribution
sns.countplot(x['region'])
plt.title("Region Distribution" ,size =12)
plt.xlabel("region", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```

```python
#boxplot for charges
sns.boxplot(x['charges'])
plt.title("Charges Distribution" ,size =12)
plt.xlabel("charges", size =12)
plt.ylabel("frequency", size =12)
plt.grid()
plt.show()
```
